```python
z_score=[]
reliance_by_apollo=reliance.Close/apollo.Close
for i in range(252,len(reliance.Close)):

    c=reliance_by_apollo.iloc[i-251:i+1]
    z_score.append((reliance_by_apollo[i]-c.mean())/(c.std()))
idbi_by_axis=(idbi.Close)/(axis.Close)
z_score2=[]
for i in range(252,len(idbi.Close)):

    c=idbi_by_axis.iloc[i-251:i+1]
    z_score2.append((idbi_by_axis[i]-c.mean())/(c.std()))
plt.plot(z_score)
plt.axhline(0.7,color='green')
plt.axhline(-0.7,color='red')
plt.axhline(0,color='black')
```

```python
def without_stop_loss(stock1,stock2,z_score,SD):
    trade=[]
    number=[] #1 for stock1 to be shorted and 2 for stock 2 to be shorted
    square_off=[]
    capital_list=[]
    count=0
    capital_for_stock1=2500000
    capital_for_stock2=2500000
    non_investment_capital=0

    for i in range(len(z_score)-1):

        if(i<=count):
            continue
        if(z_score[i]<SD and z_score[i]>-SD):
            capital=capital_for_stock1+capital_for_stock2+non_investment_capital
            capital_list.append(capital)
        if(z_score[i]>SD):
            x=len(square_off)
            trade.append(i+252)
            number.append(1)
            capital=capital_for_stock1+capital_for_stock2+non_investment_capital
            capital_list.append(capital)
            capital_for_stock1=capital/2
            capital_for_stock2=capital/2
            number_of_stock1=int(capital_for_stock1/stock1.Close[i+252])
            number_of_stock2=int(capital_for_stock2/stock2.Close[i+252])
            capital_for_stock1+=(number_of_stock1)*(stock1.Close[i+252])
            capital_for_stock2-=(number_of_stock2)*(stock2.Close[i+252])
            non_investment_capital=capital-(number_of_stock1)*(stock1.Close[i+252])-(number_of_stock2)*(stock2.Close[i+252])
            for j in range(i+1,len(z_score)):
                capital_for_stock1-=(number_of_stock1)*(stock1.Close[j])
                capital_for_stock2+=(number_of_stock2)*(stock2.Close[j])
                capital=capital_for_stock1+capital_for_stock2+non_investment_capital
                capital_list.append(capital)
                if(z_score[j]<0):
                    square_off.append(j+252)
                    count=j
                    break
                capital_for_stock1+=(number_of_stock1)*(stock1.Close[j])
                capital_for_stock2-=(number_of_stock2)*(stock2.Close[j])
            if(len(square_off)==x):
                square_off.append(len(z_score)+251)
                count=len(z_score)-1
                break
        if(z_score[i]<-SD):
            x=len(square_off)
            trade.append(i+252)
            number.append(2)
            capital=capital_for_stock1+capital_for_stock2+non_investment_capital
            capital_list.append(capital)
            capital_for_stock1=capital/2
            capital_for_stock2=capital/2
            number_of_stock1=int(capital_for_stock1/stock1.Close[i+252])
            number_of_stock2=int(capital_for_stock2/stock2.Close[i+252])
            capital_for_stock1-=(number_of_stock1)*(stock1.Close[i+252])
            capital_for_stock2+=(number_of_stock2)*(stock2.Close[i+252])
            non_investment_capital=capital-(number_of_stock1)*(stock1.Close[i+252])-(number_of_stock2)*(stock2.Close[i+252])
            for j in range(i+1,len(z_score)):
                capital_for_stock1+=(number_of_stock1)*(stock1.Close[j])
                capital_for_stock2-=(number_of_stock2)*(stock2.Close[j])
                capital=capital_for_stock1+capital_for_stock2+non_investment_capital
                capital_list.append(capital)
                if(z_score[j]>0):
                    square_off.append(j+252)
                    count=j
                    break
                capital_for_stock1-=(number_of_stock1)*(stock1.Close[j])
                capital_for_stock2+=(number_of_stock2)*(stock2.Close[j])
            if(len(square_off)==x):
                square_off.append(len(z_score)+251)
                count=len(z_score)-1
                break
    c=pd.DataFrame(columns=['trade','number','square_off'])
    c['trade']=trade
    c['number']=number
    c['square_off']=square_off
    plt.plot(capital_list,color='r')
    returns=(capital_list[-1]-capital_list[0])/(capital_list[0])
    print("% returns: ",returns*100," %")
    print("capital: ",capital_list[-1])
    print("non investment capital: ",non_investment_capital)
    return c
```

```python
def with_stop_loss(stock1,stock2,z_score,SD):
    trade=[]
    number=[]
    square_off=[]
    pointer=0
    capital_for_stock1=2500000
    capital_for_stock2=2500000
    non_investment_capital=0
    capital=0

    capital_list=[]
    for i in range(len(z_score)-1):
        if(i<pointer):
            continue
        if(z_score[i]<SD and z_score[i]>-SD):
            capital=(capital_for_stock1)+(capital_for_stock2)+non_investment_capital

            capital_list.append(capital)
        if(z_score[i]>SD):
            x=len(square_off)
            trade.append(i+252)
            number.append(1)
            capital=(capital_for_stock1)+(capital_for_stock2)+non_investment_capital
            stop_loss=(0.95)*capital
            capital_list.append(capital)
            capital_for_stock1=capital/2
            capital_for_stock2=capital/2
            number_of_stock1=int(capital_for_stock1/stock1.Close[i+252])
            number_of_stock2=int(capital_for_stock2/stock2.Close[i+252])
            capital_for_stock1+=(number_of_stock1)*(stock1.Close[i+252])
            capital_for_stock2-=(number_of_stock2)*(stock2.Close[i+252])
            non_investment_capital=capital-(number_of_stock1)*(stock1.Close[i+252])-(number_of_stock2)*(stock2.Close[i+252])
            for j in range(i+1,len(z_score)):
                capital_for_stock1-=(number_of_stock1)*(stock1.Close[j])
                capital_for_stock2+=(number_of_stock2)*(stock2.Close[j])
                capital=capital_for_stock1+capital_for_stock2+non_investment_capital
                capital_list.append(capital)
                if(capital>(1.05*(stop_loss/0.95))):
                    stop_loss=0.95*capital
                if(capital<stop_loss):
                    print("Stop loss")
                    square_off.append(j+252)
                    pointer=j
                    break
                if(z_score[j]<0):
                    square_off.append(j+252)
                    pointer=j
                    break
                capital_for_stock1+=(number_of_stock1)*(stock1.Close[j])
                capital_for_stock2-=(number_of_stock2)*(stock2.Close[j])
            if(len(square_off)==x):
                square_off.append(len(z_score)+251)
                pointer=len(square_off)
                break
        if(z_score[i]<-SD):
            x=len(square_off)
            trade.append(i+252)
            number.append(2)
            capital=(capital_for_stock1)+(capital_for_stock2)+non_investment_capital
            stop_loss=(0.95)*capital
            capital_list.append(capital)
            capital_for_stock1=capital/2
            capital_for_stock2=capital/2
            number_of_stock1=int(capital_for_stock1/stock1.Close[i+252])
            number_of_stock2=int(capital_for_stock2/stock2.Close[i+252])
            capital_for_stock1-=(number_of_stock1)*(stock1.Close[i+252])
            capital_for_stock2+=(number_of_stock2)*(stock2.Close[i+252])
            non_invested_capital=capital-(number_of_stock1)*(stock1.Close[i+252])-(number_of_stock2)*(stock2.Close[i+252])
            for j in range(i+1,len(z_score)):
                capital_for_stock1+=(number_of_stock1)*(stock1.Close[j])
                capital_for_stock2-=(number_of_stock2)*(stock2.Close[j])
                capital=capital_for_stock1+capital_for_stock2+non_investment_capital
                capital_list.append(capital)


                if(capital>(1.1*(stop_loss/0.95))):
                    stop_loss=(0.95)*capital
                if(capital<stop_loss):
                    print("Stop loss")
                    square_off.append(j+252)
                    pointer=j
                    break
                if(z_score[j]>0):
                    square_off.append(j+252)
                    pointer=j
                    break
                capital_for_stock1-=(number_of_stock1)*(stock1.Close[j])
                capital_for_stock2+=(number_of_stock2)*(stock2.Close[j])
            if(len(square_off)==x):
                square_off.append(len(z_score)+251)
                pointer=len(square_off)-1
                break

    c=pd.DataFrame(columns=['trade','number','square_off'])
    c['trade']=trade
    c['number']=number
    c['square_off']=square_off
    print(c)

    plt.plot(capital_list,color='blue')
    returns=(capital_list[-1]-capital_list[0])/(capital_list[0])
    print("% returns: ",returns*100," %")
    print("capital: ",capital_list[-1])
    print("non investment capital: ",non_investment_capital)
    return c
```
